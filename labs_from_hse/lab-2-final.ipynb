#### Задача 6

```python
import scipy.spatial
```

```python
def compute(a, b):
    m, n = a.shape
    resalting_matrix = np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            temp_dist = np.linalg.norm(a[i][j] - b[i][j])
            resalting_matrix[i][j] = temp_dist

    return resalting_matrix

a = np.random.normal(size=(10, 3))
b = np.random.normal(size=(10, 3))

compute(a, b)
```

```python
def pairplot(df, target):
    ncol, nrow = 7, df.shape[1] // 7 + (df.shape[1] % 7 > 0)
    plt.figure(figsize=(ncol * 4, nrow * 4))

    for i, feature in enumerate(df.columns):
        plt.subplot(nrow, ncol, i + 1)
        plt.scatter(df[feature], target, s=10, marker='o', alpha=.6)
        plt.xlabel(feature)
        if i % ncol == 0:
            plt.ylabel('target')
```

Для одной комбинаций метода масштабирования и линейной регрессии постройте зависимость метрики $R^2$ на кросс-валидации (`GridSearchCV.cv_results_['mean_test_score']`) от значения параметра регуляризации для различных степеней полиномов в `PolynomialFeatures`.

```python
from dataclasses import dataclass 

# Here we want to properly group data and present it in the subplots


params_out = gridsrc_ridge.cv_results_["params"]
scores = gridsrc_ridge.cv_results_["mean_test_score"]

@dataclass
class ResultsStructured:
    def __init__(self, alpha, poly_degree, score):
        self.score = score
        self.poly_degree = poly_degree
        self.alpha = alpha

results = []

for ind in range(len(params_out)):
    param = params_out[ind]
    keys = list(param.keys())
    alpha = param[keys[1]]
    poly_degree = param[keys[0]]
    score = scores[ind]
    results.append(ResultsStructured(alpha=alpha, poly_degree=poly_degree, score=score))

poly_features = []

for i in range(2, 6):
    poly_group = [result for result in results if result.poly_degree == i]
    alphas = [poly.alpha for poly in poly_group]
    scores = [poly.score for poly in poly_group]
    poly_features.append((alphas,scores))

# Alphas and scores sequences for all polynomial feature values tested

poly2_alphas, poly2_scores = poly_features[0]
poly3_alphas, poly3_scores = poly_features[1]
poly4_alphas, poly4_scores = poly_features[2]
poly5_alphas, poly5_scores = poly_features[3]

# Create subplots 

f, (line1, line2) = plt.subplots(2, 2, figsize=(8, 6))
f.suptitle("Score-Alpha dependency for Ridge with StandardScaler")

# Subplot for Polynomial feature value '2'

line1[0].semilogx(poly2_alphas, poly2_scores)
line1[0].set_title('Polynomial Feature Value "2"')
line1[0].set_ylabel('CV score')
line1[0].set_xlabel('alpha')
line1[0].axhline(np.max(poly2_scores), linestyle='--', color='.5')
line1[0].set_xlim([poly2_alphas[0], poly2_alphas[-1]])

# Subplot for Polynomial feature value '3'

line1[1].semilogx(poly3_alphas, poly3_scores)
line1[1].set_title('Polynomial Feature Value "3"')
line1[1].set_ylabel('CV score')
line1[1].set_xlabel('alpha')
line1[1].axhline(np.max(poly3_scores), linestyle='--', color='.5')
line1[1].set_xlim([poly3_alphas[0], poly3_alphas[-1]])

# Subplot for Polynomial feature value '4'

line2[0].semilogx(poly4_alphas, poly4_scores)
line2[0].set_title('Polynomial Feature Value "4"')
line2[0].set_ylabel('CV score')
line2[0].set_xlabel('alpha')
line2[0].axhline(np.max(poly4_scores), linestyle='--', color='.5')
line2[0].set_xlim([poly4_alphas[0], poly4_alphas[-1]])

# Subplot for Polynomial feature value '5'

line2[1].semilogx(poly5_alphas, poly5_scores)
line2[1].set_title('Polynomial Feature Value "5"')
line2[1].set_ylabel('CV score')
line2[1].set_xlabel('alpha')
line2[1].axhline(np.max(poly5_scores), linestyle='--', color='.5')
line2[1].set_xlim([poly5_alphas[0], poly5_alphas[-1]])

# Show
f.tight_layout()
f.show()
```

При каких значениях происходит переобучение? А недообучение? Почему?

__Ответ:__

Недообучение, происходит при высоких значениях alpha, когда штраф, накладываемый на модель становится слишком высоким и модель приближается к константе. Переобучение происходит прпи низких альфа - когда незначительные features недостаточно штрафуются и в итоге имеют слишком высокий вес в модели, что приводит к тому, что модель цепляется за незначительные данные, а это в свою очередь не позволяет ей удачно генерализоваться.
